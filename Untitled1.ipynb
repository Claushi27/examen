```python
# Definición de los precios de los departamentos
precios_venta = {
    'A': 3800,
    'B': 3000,
    'C': 2800,
    'D': 3500,
}

# Definición de los departamentos disponibles
pisos = 10
departamentos_por_piso = 4

departamentos_disponibles = [[True] * departamentos_por_piso for _ in range(pisos)]

# Definición de la lista de compradores
compradores = []

# Función para mostrar el menú y obtener la opción seleccionada por el usuario
def mostrar_menu():
    print("----- Menú -----")
    print("1. Comprar departamento")
    print("2. Mostrar departamentos disponibles")
    print("3. Ver listado de compradores")
    print("4. Mostrar ganancias totales")
    print("5. Salir")
    opcion = input("Seleccione una opción: ")
    return opcion

def comprar_o_arrendar_departamento():
    print("----- Comprar/Arrendar departamento -----")
    piso = int(input("Ingrese el número de piso (1-10): "))
    if not (1 <= piso <= pisos):
        print("Piso inválido.")
        return

    departamentos_piso = ['A', 'B', 'C', 'D']
    tipo_departamento = input(f"Ingrese el tipo de departamento {departamentos_piso} (A-D): ").upper()
    if tipo_departamento not in departamentos_piso:
        print("Tipo de departamento inválido.")
        return

    departamento = tipo_departamento + str(piso)

    print("1. Comprar")
    print("2. volver al menu anterior")
    opcion = input("Seleccione una opción: ")

    if opcion == '1':
        if not departamentos_disponibles[piso - 1][ord(tipo_departamento) - ord('A')]:
            print("Departamento no disponible.")
            return
        run = input("Ingrese el RUN del comprador: ")
        departamentos_disponibles[piso - 1][ord(tipo_departamento) - ord('A')] = False
        compradores.append({'run': run, 'departamento': departamento})
        print("Compra realizada correctamente.")
    elif opcion == '2':
         return
    else:
        print("Opción inválida. Por favor, seleccione una opción válida.")

def mostrar_departamentos_disponibles():
    print("----- Departamentos disponibles -----")
    for piso, departamentos in enumerate(departamentos_disponibles, start=1):
        print(f"Piso {piso}: ", end="")
        departamentos_piso = ['A', 'B', 'C', 'D', 'E']
        for i, disponible in enumerate(departamentos):
            tipo_departamento = departamentos_piso[i]
            if disponible:
                print(f"{tipo_departamento}{piso} (Disponible) ", end="")
            else:
                print(f"{tipo_departamento}{piso} (No disponible) ", end="")
        print()


# Función para mostrar el listado de compradores
def mostrar_listado_compradores():
    print("----- Listado de compradores -----")
    compradores_ordenados = sorted(compradores, key=lambda x: x['run'])
    for comprador in compradores_ordenados:
        print(f"RUN: {comprador['run']}, Departamento: {comprador['departamento']}")

# Función para mostrar las ganancias totales
def mostrar_ganancias_totales():
    print("----- Ganancias totales -----")
    total_ventas = 0
    for departamento in compradores:
        tipo_departamento = departamento['departamento'][0]
        if tipo_departamento in precios_venta and 'Arriendo' not in departamento['departamento']:
            total_ventas += precios_venta[tipo_departamento]

    print(f"Ventas totales: ${total_ventas}")

# Función principal del programa
def main():
    while True:
        opcion = mostrar_menu()

        if opcion == '1':
            comprar_o_arrendar_departamento()
        elif opcion == '2':
            mostrar_departamentos_disponibles()
        elif opcion == '3':
            mostrar_listado_compradores()
        elif opcion == '4':
            mostrar_ganancias_totales()
        elif opcion == '5':
            print(f"¡Gracias por utilizar el sistema, de claudio cabrera!")
            break
        else:
            print("Opción inválida. Por favor, seleccione una opción válida.")

# Ejecución del programa
if __name__ == '__main__':
    main()
```
